```python
for i in ['MSSubClass', 'LotFrontage', 'LotArea', 'OverallQual', 'OverallCond',
       'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2',
       'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
       'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
       'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
       'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF',
       'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal',
       'MoSold', 'YrSold'] :
    train[i]=train[i].fillna(train[i].mean())
    test[i]=test[i].fillna(test[i].mean())
```

```python
#Inserting the null values with 'None' for columns containing non-numeric data type
```

```python
for i in ['MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
       'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
       'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
       'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
       'Functional', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
       'PavedDrive', 'SaleType', 'SaleCondition']:
    train[i]=train[i].fillna('None')
    test[i]=test[i].fillna('None')
```

```python
#Creating a new column 'train' in the dataframe in which '1' is assigned to the training data and '0' to the test data. 
#Subsequently applying One-Hot encoding as our data contains a number of categorical data types. 
#The new column has been created to apply One-Hot encoding simultaneously to both the train and test data. 
#The new column has been created to easily separate the train and test data after the application of One-Hot encoding.
```

```python
train['train']=1
test['train']=0
DF=pd.concat([train,test])
```

```python
new_df=pd.get_dummies(DF,columns=['MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
       'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
       'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
       'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
       'Functional', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
       'PavedDrive', 'SaleType', 'SaleCondition'],drop_first=True)
```

```python
#Segregating back the data into traing and test after One hot encoding 
```

```python
train=new_df[new_df['train']==1]
test=new_df[new_df['train']==0]
```

```python
train.head()
```

```python
test.head()
```

```python
#dropping 'train' column from train dataframe and 'train' & 'Saleprice' from test dataframe
```

```python
train=train.drop(['train'],axis=1)
test=test.drop(['train','SalePrice'],axis=1)
```

```python
#Creating inputs for model application
```

```python
x_train=train.drop('SalePrice',axis=1)
y_train=train['SalePrice']
x_test=test
```

```python
#Applying Linear Regression to our Regression Problem
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
#Creating a submission dataframe and finally saving our submission .csv file 
```

```python
d={'ID': test_parent['Id'], 'SalePrice': y_pred}
submission=pd.DataFrame(d)
print(submission)
submission.to_csv('house_submission.csv')
```
